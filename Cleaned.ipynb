```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns  
```

```python
df = pd.read_csv('/content/Task 3_Final Content Data set.csv')
```

```python
df
```

```python
df['time'] = pd.to_datetime(df['Datetime'])

```

```python
df.set_index('time', inplace=True)
```

```python
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df
```

```python
# group the DataFrame by "Content Type" and get the count
content_type_count = df.groupby('Content Type').count()
```

```python
content_type_count
```

```python
plt.pie(content_type_count['Content ID'], labels=content_type_count.index, autopct='%1.1f%%')
```

```python
Category_count = df.groupby('Category').count()
```

```python
Category_count
```

```python
plt.figure(figsize=(10, 6))
plt.bar(Category_count.index, Category_count['Content ID'])
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Number of Posts')
plt.title('Number of Posts by Category')
plt.show()
```
